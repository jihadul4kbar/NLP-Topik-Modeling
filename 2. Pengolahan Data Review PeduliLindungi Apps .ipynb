1. Menghapus Tanda Baca

```python
import string
import re

def remove_punct(text):
    text = "".join([karakter for karakter in text if karakter not in string.punctuation])
    text = re.sub('[0-9]+','',text)
    return text
```

```python
review_data['ReviewPunct'] = review_data['content'].apply(lambda x: remove_punct(x))
review_data[['content', 'ReviewPunct']]
```

Menghapus karater kosong

```python
# menghapus spasi
def remove_whitespace(text):
    return text.strip()
```

```python
review_data['ReviewRemoveWhitespace'] = review_data['ReviewPunct'].apply(lambda x: remove_whitespace(x))
review_data[['ReviewPunct','ReviewRemoveWhitespace']]
```

```python
#menghapus multipel spasi
def remove_whitespace_multiple(text):
    return re.sub('\s',' ', text)
```

```python
review_data['ReviewRemoveDoubleWhitespace'] = review_data['ReviewPunct'].apply(lambda x: remove_whitespace(x))
review_data[['ReviewPunct','ReviewRemoveDoubleWhitespace']]
```

2. Tokenizing dan Merubah Menjadi Huruf Kecil

```python
def tokenization(text):
    text = re.split('\W+', text)
    return text
```

```python
review_data['ReviewTokenization'] = review_data['ReviewPunct'].apply(lambda x: tokenization(x.lower()))
review_data[['ReviewPunct','ReviewTokenization']]
```

3. Melakukan Stop Word Bahasa Indonesia

```python
import nltk 
from nltk.tokenize import word_tokenize

nltk.download('stopwords')
```

```python
stopword = nltk.corpus.stopwords.words('indonesian')
stopword[:20]
```

```python
def hapus_stopwords(text):
    text = [kata for kata in text if kata not in stopword]
    return text
```

```python
review_data['ReviewStopword'] = review_data['ReviewTokenization'].apply(lambda x: hapus_stopwords(x))
review_data[['ReviewTokenization','ReviewStopword']]
```

4. Analisis Kata Pendek

```python
kata_pendek = []

for row in review_data['ReviewStopword']:
    for word in row:
        if len(word) <= 3:
            kata_pendek.append(word)
print(kata_pendek)
len(kata_pendek)
```

```python
import collections
ambil=collections.Counter(kata_pendek).most_common()
for kata in ambil:
    print("%s\t: %d"%(kata[0],kata[1]))
```
